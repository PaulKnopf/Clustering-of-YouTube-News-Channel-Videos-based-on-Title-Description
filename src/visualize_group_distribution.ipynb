```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
from datasets import Dataset
```

```python
colors_set = ['#40512D', '#5C733F', '#789451', '#A5A553', '#FFC857', '#E96356', '#714C3D', '#965369', '#B171CE', '#523646', '#793CEA']
```

```python
# load data
videos = Dataset.load_from_disk('../data/videos')
# load labels
labels = np.load('../data/cluster_labels.npy')
```

```python
df = pd.DataFrame(videos)
df['label'] = labels
```

```python
seaborn.set()
ag = df.groupby(['channel','label']).sum().unstack()
ag.columns = ag.columns.droplevel()

ag.plot(kind='bar',
        # colormap=cm.Accent,
        width=1,
        figsize=(16,16),
        color=colors_set,
        stacked=True
        )
plt.show()
```
