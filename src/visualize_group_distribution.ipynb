```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn
from datasets import Dataset
import spacy
```

```python
# load data
videos = Dataset.load_from_disk('../data/videos')
# load labels
labels = np.load('../data/cluster_labels.npy')
```

```python
df = pd.DataFrame(videos)
df['label'] = labels
```

```python
df_sorted = df.sort_values(by='label')
```

```python
df_sorted
```

```python
seaborn.set() #make the plots look pretty

ax = df.groupby(['channel','label']).sum().unstack()
ax.columns = ax.columns.droplevel()

ax.plot(kind = 'bar', colormap = cm.Accent, width = 1, figsize=(16,16))
plt.show()
```
